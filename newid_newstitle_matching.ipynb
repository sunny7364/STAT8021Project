```python
#这里是对相似用户的结果做匹配

import pandas as pd
import numpy as np
import json

newsinfo_trainingdataset = pd.read_csv('news.tsv', sep = '\t', header = None, index_col = None, names=['News ID', 'Category', 'SubCategory', 'Title', 'Abstract', 'URL', 'Title Entities', 'Abstract Entites'])
#print(newsinfo_trainingdataset）
newsinfo_testingdataset = pd.read_csv('news_test.tsv', sep = '\t', header = None, index_col = None, names=['News ID', 'Category', 'SubCategory', 'Title', 'Abstract', 'URL', 'Title Entities', 'Abstract Entites'])
newsinfo_validdataset = pd.read_csv('news_valid.tsv', sep = '\t', header = None, index_col = None, names=['News ID', 'Category', 'SubCategory', 'Title', 'Abstract', 'URL', 'Title Entities', 'Abstract Entites'])
newsinfo = pd.concat([newsinfo_trainingdataset, newsinfo_testingdataset, newsinfo_validdataset])
newsinfo.drop_duplicates()

user_simi_user_with_title = []

with open('user_simi_user.json', 'r') as similar_user_file:
    for line in similar_user_file:
        linedict = similar_user_file.readline()
        similar_user = json.loads(linedict)
        print(type(similar_user))
        print(similar_user, '\n改写前dict')
        for key, value in similar_user.items():
            if key == 'items':
                for x in value:#长度等于历史浏览过的新闻的个数
                    title = str(newsinfo.loc[newsinfo['News ID'] == 'N' + str(x[0]), 'Title'].values[0])
                    # title.replace("['", "")
                    # title.replace("']", "")
                    # title.replace("[\"", "")
                    # title.replace("\"]", "")
                    # title.replace("'\n", "")
                    # title.replace("\"\n", "")
                    x.insert(1, title)
            if key == 'similar_users':
                for j in range(len(value)):#长度是相似用户的个数
                    for y in value[j]['items']:
                        title = str(newsinfo.loc[newsinfo['News ID'] == 'N' + str(y[0]), 'Title'].values[0])
                        y.insert(1, title)
        print(similar_user, '\n改写后dict')
        user_simi_user_with_title.append(similar_user)

with open("user_simi_user_with_newstitle.json", "w") as f:
    for line in user_simi_user_with_title:
        f.write(json.dumps(line)+"\n")
```
